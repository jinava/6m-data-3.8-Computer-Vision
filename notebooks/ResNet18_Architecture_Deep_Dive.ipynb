# Deep Dive: Demystifying the Black Box

This notebook breaks open the pre-trained `ResNet18` model to physically look at its architecture, its hyperparameters, and its raw mathematical weights and biases!

<img src="resnet18_architecture.png" width="400"/>

ResNet18 has 18 Layers:
- Initial Convolution: 1 layer (conv1)
- Layer 1: 2 BasicBlocks × 2 Conv layers each (conv1 + conv2) = 4 layers
- Layer 2: 2 BasicBlocks × 2 Conv layers each (conv1 + conv2) = 4 layers
- Layer 3: 2 BasicBlocks × 2 Conv layers each (conv1 + conv2) = 4 layers
- Layer 4: 2 BasicBlocks × 2 Conv layers each (conv1 + conv2) = 4 layers
- Fully Connected: 1 layer (fc)

Total: 1 + 4 + 4 + 4 + 4 + 1 = 18 layers.

## 1. The "Big Picture" Architecture
The easiest way to see the hyperparameters (kernel size, stride, padding) is to print the model directly.

````python
import os
## You need to set this environment variable for running the codes on Mac!!! Otherwise,
## your kernel may crash!!!
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"

import torchvision.models as models

# 1. Hire the "College Graduate" (Load ResNet18)
resnet = models.resnet18(weights='DEFAULT')

# 2. Print the entire model architecture
print(resnet)
````

## 2. Extracting ALL Hyperparameters in Code
If you want to programmatically extract all settings (kernels, strides, padding, biases, etc.), you can loop through every layer.

````python
import torch.nn as nn

print("--- ANALYZING RESNET18 ARCHITECTURE ---")

for name, layer in resnet.named_modules():
    
    if isinstance(layer, nn.Conv2d):
        print(f"[{name}] CONVOLUTION LAYER:")
        print(f"  - In Channels: {layer.in_channels}")
        print(f"  - Out Channels (Num Kernels): {layer.out_channels}")
        print(f"  - Kernel Size: {layer.kernel_size}")
        print(f"  - Stride: {layer.stride}")
        print(f"  - Padding: {layer.padding}")
        print(f"  - Has Bias?: {layer.bias is not None}\n")
        
    elif isinstance(layer, (nn.MaxPool2d, nn.AdaptiveAvgPool2d)):
        print(f"[{name}] POOLING LAYER:")
        if hasattr(layer, 'kernel_size'):
            print(f"  - Pool Size (Kernel): {layer.kernel_size}")
        if hasattr(layer, 'stride'):
            print(f"  - Stride: {layer.stride}")
        if hasattr(layer, 'padding'):
            print(f"  - Padding: {layer.padding}")
        print("\n")
        
    elif isinstance(layer, nn.Linear):
        print(f"[{name}] LINEAR CLASSIFIER HEAD:")
        print(f"  - Input Neurons: {layer.in_features}")
        print(f"  - Output Predictions: {layer.out_features}")
        print(f"  - Has Bias?: {layer.bias is not None}\n")
````

## 3. Looking at the Conv1 Weights (All 3 Channels)
Let's extract the raw decimal numbers for the very first 7x7 edge detector. Because input images have 3 color channels (RGB), this single edge detector is actually made of three 7x7 grids!

Change the codes (e.g. resnet.conv2) to view the raw numbers of other layers (e.g. resnet.conv2 to view info of Conv Layer 2)

````python
print("--- LOOKING INSIDE THE RESNET18 FOR LEARNT WEIGHTS ---")

layer_of_interest = resnet.conv1
## Some sample codes here for your reference:
## layer_of_interest = resnet.layer2[0].conv2 # Second conv layer in first basic block of ResNet Layer 2.
## layer_of_interest = resnet.layer1[1].conv1 # First conv layer in second basic block of ResNet Layer 1.

weights = layer_of_interest.weight.data
print(f"Weights Shape: {weights.shape}")
# Shape is (64, 3, 7, 7)
# Meaning: 64 Kernels, looking at 3 RGB channels, using a 7x7 grid.

print(f"\n--- KERNEL 0: RED CHANNEL (7x7) ---")
print(weights[0, 0, :, :])

print(f"\n--- KERNEL 0: GREEN CHANNEL (7x7) ---")
print(weights[0, 1, :, :])

print(f"\n--- KERNEL 0: BLUE CHANNEL (7x7) ---")
print(weights[0, 2, :, :])
````

## 4. The Fully Connected (FC) Layer
How many hidden layers are in ResNet18`s Fully Connected section? **Zero!**

In classic MLPs (like the one we built on Monday), we had multiple hidden FC layers. But in modern CNNs, 99% of the heavy lifting is done by the Convolutional Blocks! The FC section at the very end is literally just **one single linear layer** acting as the final voting head to classify the 1000 categories.

### Where do the 512 inputs come from?
Right before the data hits this FC layer, it goes through **Global Average Pooling**.
1. The final convolutional block (Layer 4) outputs 512 different "feature maps" (heatmaps).
2. The Global Average Pooling layer mathematically squashes each of those 512 heatmaps down into one single average number.
3. So, the massive 3D image data is completely flattened into a simple 1D array of exactly **512 numbers**.

Those 512 numbers are the direct inputs to this FC layer. That is why the weight matrix shape is `(1000, 512)`: **1000 output classes × 512 input numbers.**

Now that we know exactly where the inputs come from, let`s print out the actual voting weights in the code below!

````python
print("--- FULLY CONNECTED (FC) LAYER ANALYSIS ---")
fc_layer = resnet.fc

# 1. Look at the FC WEIGHTS
fc_weights = fc_layer.weight.data
print(f"FC Weights Shape: {fc_weights.shape}") 
# Shape is (1000, 512)
# Meaning: 1000 output classes, each looking at 512 input features from the backbone.

print("\nHere are the first 10 weights connecting the 512 backbone features to Class 0:")
print(fc_weights[0, :10])

# 2. Look at the FC BIASES
fc_biases = fc_layer.bias.data
print(f"\nFC Biases Shape: {fc_biases.shape}")
# Shape is (1000)
# Meaning: 1 single bias number added to each of the 1000 output categories.

print("\nHere are the bias numbers for the 1000 output categories:")
print(fc_biases[:])
````

## 7. Cheat Sheet: Mapping `torchinfo` back to PyTorch Architecture
To help cross-reference the `torchinfo` indexing numbers back to the original PyTorch model layer names (which we printed in Cell 1), here is the exact translation of the main components:

- **`Conv2d: 1-1`** $\rightarrow$ `conv1` (Initial 7x7 Convolution)
- **`BatchNorm2d: 1-2`** $\rightarrow$ `bn1` (Initial Batch Normalization)
- **`ReLU: 1-3`** $\rightarrow$ `relu` (Initial Activation)
- **`MaxPool2d: 1-4`** $\rightarrow$ `maxpool` (Initial Pooling)
- **`Sequential: 1-5`** $\rightarrow$ `layer1` (The 1st Macro-Block of 4 Conv layers)
- **`Sequential: 1-6`** $\rightarrow$ `layer2` (The 2nd Macro-Block of 4 Conv layers)
- **`Sequential: 1-7`** $\rightarrow$ `layer3` (The 3rd Macro-Block of 4 Conv layers)
- **`Sequential: 1-8`** $\rightarrow$ `layer4` (The 4th Macro-Block of 4 Conv layers)
- **`AdaptiveAvgPool2d: 1-9`** $\rightarrow$ `avgpool` (The Global Average Pooling layer)
- **`Linear: 1-10`** $\rightarrow$ `fc` (The Final Fully Connected Classification Head)

## Priority 1 — The Residual Connection (the "Res" in ResNet)

**<span style="color:green">[Opus 4.8]</span> The single most important idea in the architecture:**

Every `BasicBlock` doesn't just transform its input `x` — it computes `F(x)` (the conv→BN→ReLU→conv→BN path) and then **adds the original input back**:

$$\text{out} = \text{ReLU}\big(F(x) + x\big)$$

That `+ x` is the **skip connection** (a.k.a. shortcut / identity). It is why an 18-layer — or even a 152-layer — network trains at all: gradients can flow straight through the `+ x` highway, sidestepping the vanishing-gradient problem that stalled deep *plain* networks. The block only has to learn the *residual* (the change) `F(x)`, not the entire mapping — hence "**Res**idual Network."

**The `downsample` branch:** when a stage changes channel count or spatial size (e.g. 64→128 with stride 2), `x` no longer matches the shape of `F(x)`, so they can't be added directly. The `1×1` `downsample` conv reshapes `x` so the addition lines up. That is its *only* job — it is plumbing for the shortcut, which is why the standard "18-layer" count ignores it.

**<span style="color:green"><em>[Opus 4.8] — end of note</em></span>

````python
import torch, inspect
import torchvision.models as models

# Print the ACTUAL forward() of a BasicBlock, straight from torchvision's source:
print(inspect.getsource(models.resnet.BasicBlock.forward))

# Now prove "out = ReLU(F(x) + x)" numerically on the first block of layer1.
# (layer1 has no downsample, so the shortcut 'identity' is the raw input x.)
resnet.eval()  # eval mode -> BatchNorm uses fixed running stats, so the result is deterministic

block = resnet.layer1[0]
x = torch.randn(1, 64, 56, 56)

# F(x): the conv -> bn -> relu -> conv -> bn path
F_x = block.bn2(block.conv2(block.relu(block.bn1(block.conv1(x)))))
identity = x                          # <-- the skip connection
manual = block.relu(F_x + identity)   # out = ReLU(F(x) + x)

real = block(x)                       # what the block actually computes internally
print("\nReproduced block() by hand with the '+ x' shortcut? ->",
      torch.allclose(manual, real, atol=1e-6))
````

## Priority 2 — Real Feature Maps (not a hand-typed one)

**<span style="color:green">[Opus 4.8]</span> Replacing the illustrative tensor in Section 5 with the genuine article:**

Section 5 *described* a 7×7 heatmap using hand-typed numbers. Here we capture the **actual** activations by registering a **forward hook** on `layer4`, pushing a real photo through, and plotting the strongest channels. Each of the 512 channels is a 7×7 map; bright regions = "this feature fired here." This is the real internal state of the network on a real image — swap in your own image path to explore.

**<span style="color:green"><em>[Opus 4.8] — end of note</em></span>

````python
import torch, urllib.request
from PIL import Image
import matplotlib.pyplot as plt
import torchvision.models as models

weights = models.ResNet18_Weights.DEFAULT   # the same DEFAULT weights loaded in Section 1
preprocess = weights.transforms()            # official resize / center-crop / normalize pipeline

# Grab a sample image (any RGB photo works — point this at your own file instead)
urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/pytorch/hub/master/images/dog.jpg", "sample.jpg")
img = Image.open("sample.jpg").convert("RGB")
batch = preprocess(img).unsqueeze(0)         # shape (1, 3, 224, 224)

# A forward hook grabs layer4's output as the image flows through the network
activations = {}
handle = resnet.layer4.register_forward_hook(
    lambda module, inp, out: activations.__setitem__("layer4", out.detach()))

resnet.eval()
with torch.no_grad():
    resnet(batch)
handle.remove()

feature_maps = activations["layer4"][0]      # (512, 7, 7)  <- reused by Priorities 5 & 6
print("Real layer4 feature-map stack:", tuple(feature_maps.shape))

# Show the 6 most-activated channels for this image
strongest = torch.topk(feature_maps.mean(dim=(1, 2)), 6).indices
fig, axes = plt.subplots(1, 6, figsize=(12, 2.2))
for ax, ch in zip(axes, strongest):
    ax.imshow(feature_maps[ch], cmap="viridis")
    ax.set_title(f"channel {ch.item()}"); ax.axis("off")
fig.suptitle("Actual 7x7 feature maps from layer4 (most-activated channels)")
plt.show()
````

## Priority 3 — Seeing the `conv1` Filters as Images

**<span style="color:green">[Opus 4.8]</span> Turning Section 3's decimal grids into pictures:**

Section 3 printed the raw 7×7 numbers of `conv1`. Those same 64 filters become intuitive the instant you *render* them: normalize each `3×7×7` kernel to `[0, 1]` and show it as a tiny RGB tile. You'll see the network learned classic low-level detectors — **oriented edge bars** (black/white stripes at various angles) and **colour-opponent blobs** (e.g. orange-vs-blue) — entirely from data. This is the famous first-layer visualization that dates back to AlexNet.

**<span style="color:green"><em>[Opus 4.8] — end of note</em></span>

````python
import matplotlib.pyplot as plt

filters = resnet.conv1.weight.data.clone()           # (64, 3, 7, 7)
# Normalize the whole bank to [0, 1] so edges and colours are visible
filters = (filters - filters.min()) / (filters.max() - filters.min())

fig, axes = plt.subplots(8, 8, figsize=(8, 8))
for i, ax in enumerate(axes.flat):
    ax.imshow(filters[i].permute(1, 2, 0))           # (7, 7, 3) for imshow
    ax.axis("off")
fig.suptitle("All 64 learned conv1 filters (7x7 RGB) — edge bars & colour blobs")
plt.show()
````

## Priority 4 — Why `bias=False` on Convs but `bias=True` on BN & FC

**<span style="color:green">[Opus 4.8]</span> Explaining the `Has Bias?` pattern from Section 2:**

The loop in Section 2 printed `Has Bias?: False` for every convolution and `True` for the final linear layer. That isn't arbitrary:

**Why every conv has `bias=False`:** in ResNet each `Conv2d` is *immediately followed by* a `BatchNorm2d`. BatchNorm first **subtracts the per-channel mean** of its input — which mathematically *cancels any constant the conv's bias could have added* — and then applies its own learnable shift `β` (`bn.bias`). So a conv bias would be redundant: erased by BN's mean-subtraction and re-supplied by `β`. Dropping it saves parameters with **zero** loss of expressiveness. (This includes the `1×1 downsample` convs — they're followed by BN too.)

**Why `bias=True` survives in exactly two places:**
- **The BatchNorm layers themselves** — `bn.bias` *is* the shift `β`; it is the bias the conv handed off to. The bias didn't disappear, it moved into BN.
- **The final `fc` layer** — it is **not** followed by a BatchNorm. It's the last operation before the logits, so it needs its own offset to position each class score; there is nothing downstream to absorb a missing bias.

**Rule of thumb:** `Conv → BN` ⇒ turn the conv bias off (BN's `β` does the job). A layer with nothing normalizing after it (the classifier head) keeps its bias.

**<span style="color:green"><em>[Opus 4.8] — end of note</em></span>

````python
# The bias lives wherever there is NO BatchNorm downstream to absorb it.
print("conv1.bias                  :", resnet.conv1.bias)                     # None  (bn1 follows)
print("layer4[0].downsample[0].bias:", resnet.layer4[0].downsample[0].bias)  # None  (a BN follows)
print("bn1.bias  (the shift beta)  :", tuple(resnet.bn1.bias.shape), "<- the conv's bias moved here")
print("fc.bias                     :", tuple(resnet.fc.bias.shape), "<- kept: nothing normalizes after it")
````

## Priority 5 — Close the Loop: Actually Classify an Image

**<span style="color:green">[Opus 4.8]</span> Using the black box, not just dissecting it:**

We've taken the model apart; now let's *run* it. Below we feed the same photo through the full network and decode the 1000 logits into human-readable ImageNet labels (top-5 with confidences).

**Critical practical gotcha:** call `resnet.eval()` **before** inference. In `train()` mode BatchNorm normalizes using the *current batch's* statistics; with a batch of one image that yields wrong, unstable predictions. `eval()` switches BN to its stored `running_mean` / `running_var`. Forgetting this is one of the most common real-world inference bugs.

**<span style="color:green"><em>[Opus 4.8] — end of note</em></span>

````python
import torch
import torchvision.models as models

weights = models.ResNet18_Weights.DEFAULT
categories = weights.meta["categories"]   # 1000 human-readable ImageNet class names

resnet.eval()                             # <-- essential: BN uses running stats, not batch stats
with torch.no_grad():
    logits = resnet(batch)                # 'batch' = the preprocessed image from Priority 2
probs = logits.softmax(dim=1)[0]

top5 = torch.topk(probs, 5)
print("Top-5 predictions:")
for score, idx in zip(top5.values, top5.indices):
    print(f"  {categories[idx]:<25} {score.item() * 100:5.1f}%")
````

## Priority 6 — Global Average Pooling, Shown Numerically

**<span style="color:green">[Opus 4.8]</span> Proving the "512 numbers" claim from Section 4:**

Section 4 said `avgpool` squashes the `(512, 7, 7)` stack into exactly **512 numbers**, one per channel. That operation is nothing more than the **mean of each 7×7 map**. The cell below confirms `AdaptiveAvgPool2d` is identical to `feature_maps.mean()` over the spatial dimensions — and those 512 means are precisely the vector handed to the `fc` classifier.

**<span style="color:green"><em>[Opus 4.8] — end of note</em></span>

````python
import torch
import torch.nn.functional as F

fmaps = feature_maps.unsqueeze(0)                  # (1, 512, 7, 7) captured in Priority 2
gap = F.adaptive_avg_pool2d(fmaps, 1)              # exactly what resnet.avgpool does -> (1, 512, 1, 1)
manual = fmaps.mean(dim=(2, 3), keepdim=True)      # the mean over each 7x7 grid, computed by hand

print("avgpool output shape       :", tuple(gap.shape))
print("Identical to spatial mean? :", torch.allclose(gap, manual, atol=1e-6))
print("First 5 of the 512-vector  :", [round(v, 3) for v in gap.flatten()[:5].tolist()])
print("\nThese 512 numbers are exactly what flows into the fc (Linear) classifier.")
````
